YOLOv8 _segmentation_ models use the `-seg` suffix, i.e. `yolov8n-seg.pt` and are pretrained on COCO. See [Segmentation Docs](https://docs.ultralytics.com/tasks/segment/) for full details.

```
# Load YOLOv8n-seg, train it on COCO128-seg for 3 epochs and predict an image with it
from ultralytics import YOLO

model = YOLO('yolov8m-seg.pt')  # load a pretrained YOLOv8n segmentation model
model.train(data='/content/openfield_rat/openfield_rat.yaml', epochs=3)  # train the model
model('rata.jpg')  # predict on an image
```

```
#Predicciones con imagen y labels
#Modificaciones: 
#/usr/local/lib/python3.9/dist-packages/ultralytics/yolo/engine/model.py
predict=model.predict("/content/gdrive/MyDrive/Ratainador/S10D5R29_15step_total.mp4",save=True,save_txt=True)
```

```
# Save all the labels in a single txt file
import os
path = "/content/runs/segment/predict/labels/"
filename ="S10D5R29_15step_total_"
new_path="/content/gdrive/MyDrive/Ratainador/"
seg_labels = open(new_path+filename+"labels.txt", "w")# create/open de writing file
for n_file in range(1,1542):
  if os.path.isfile(path+filename+str(n_file)+".txt")==True:#the labels are not
                                                    #in order. Check the existance
    seg_file = open(path+filename+str(n_file)+".txt", "r") #open the labels
    FileContent = seg_file.read() #read the labels
    seg_labels.write(FileContent) #group the labels in the writing file
    #print(FileContent)
```

## 3. Classification

YOLOv8 _classification_ models use the `-cls` suffix, i.e. `yolov8n-cls.pt` and are pretrained on ImageNet. See [Classification Docs](https://docs.ultralytics.com/tasks/classify/) for full details.


```
# Load YOLOv8n-cls, train it on mnist160 for 3 epochs and predict an image with it
from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt')  # load a pretrained YOLOv8n classification model
model.train(data='mnist160', epochs=3)  # train the model
model('https://ultralytics.com/images/bus.jpg')  # predict on an image
```
